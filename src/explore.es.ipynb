```python
import requests
import time

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36"
}
response = requests.get(url, headers=headers)
time.sleep(10)  # Espera 10 segundos para evitar bloqueos
html_data = response.text

if "403 Forbidden" in html_data:
    print("Acceso denegado. Intenta conectarte como usuario anónimo.")
else:
    print("Acceso exitoso.")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_data,"html.parser")
soup
```

```python
tables = soup.find_all("table")
tables
```

```python
import pandas as pd

for index, table in enumerate(tables):
    if ("Tesla Quarterly Revenue" in str(table)):
        table_index = index
        break

# Create a DataFrame
tesla_revenue = pd.DataFrame(columns = ["Date", "Revenue"])
for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        Date = col[0].text
        Revenue = col[1].text.replace("$", "").replace(",", "")
        tesla_revenue = pd.concat([tesla_revenue, pd.DataFrame({
            "Date": Date,
            "Revenue": Revenue
        }, index = [0])], ignore_index = True)

tesla_revenue.head()
```

```python
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
tesla_revenue.head()
```

```python
import sqlite3

connection = sqlite3.connect("Tesla.db")
connection
```

```python
cursor = connection.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS revenue (
    Date TEXT,
    Revenue REAL
)
""")
               
```

```python

tesla_tuples = list(tesla_revenue.to_records(index = False))
tesla_tuples[:5]
```

```python
cursor.executemany("INSERT INTO revenue VALUES (?,?)", tesla_tuples)
connection.commit()
```

```python
for row in cursor.execute("SELECT * FROM revenue"):
    print(row)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(figsize = (10, 5))

tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype('int')
sns.lineplot(data = tesla_revenue, x = "Date", y = "Revenue")

plt.tight_layout()

plt.show()
```

```python
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

tesla_revenue_yearly = tesla_revenue.groupby(tesla_revenue["Date"].dt.year)["Revenue"].sum().reset_index()

tesla_revenue_yearly.rename(columns={"Date": "Year"}, inplace=True)

tesla_revenue_filtered = tesla_revenue_yearly[tesla_revenue_yearly["Year"] < 2023]

fig, axis = plt.subplots(figsize=(10, 5))
sns.barplot(data=tesla_revenue_filtered, x="Year", y="Revenue", ax=axis)

plt.tight_layout()
plt.title("Tesla Revenue by Year")
plt.xlabel("Year")
plt.ylabel("Revenue (in millions)")
plt.show()
```

```python
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

tesla_revenue_monthly = tesla_revenue.groupby(tesla_revenue["Date"].dt.month)["Revenue"].sum().reset_index()

tesla_revenue_monthly.rename(columns={"Date": "Month"}, inplace=True)


fig, axis = plt.subplots(figsize=(10, 5))
sns.barplot(data=tesla_revenue_monthly, x="Month", y="Revenue", ax=axis)

plt.title("Tesla Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue (in millions)")
plt.tight_layout()
plt.show()
```
